# 3. Plot a bar graph of male vs female buyers participated in the sales


```python
data['Buyer Gender'].value_counts()
```

```python
data['Buyer Gender'].value_counts().plot(kind='bar')
plt.title("Male vs Female buyers participated in the sales", y=1.05);
plt.xlabel("Male vs Female Buyers", labelpad=14, color='g')
plt.ylabel("Count of People Participation", labelpad=14, color='r')
plt.show()
```

# 4. Find the top 5 cars based on their sales price

```python
dat = pd.DataFrame(data)
dat = dat[["Make","Model","Country", "City", "New Car", "Purchase Date", "Sale Price"]]
dat.nlargest(5, ['Sale Price'])
```

# 5. Find the least 5 cars based on their Resell price

```python
dat = pd.DataFrame(data)
dat = dat[["Make","Model","Country", "City", "New Car", "Purchase Date", "Resell Price"]]
dat.nsmallest(5, ['Resell Price'])
```
